```python
#Bu kod, Fizy- Müzik İkizini Bul uygulamasına ait veri setini oluşturmak için hazırlanmıştır.

import pandas as pd
import random

#www.kaggle.com sitesinden alınan kişi ve şarkı adları yüklenir.
AllSongs=pd.read_csv("train.csv")
AllSongs=AllSongs[AllSongs["explicit"]==False][["artists","track_name","track_genre","popularity"]]
AllSongs["Name"]=AllSongs["track_name"]+" | "+AllSongs["artists"]
del AllSongs["track_name"]
del AllSongs["artists"]
AllSongs = AllSongs.drop_duplicates(subset=['Name'], keep='first')

#7 adet müzik türünden en popüler 50 şarkı seçilerek veri seti oluşturulur
Acoustic=AllSongs[AllSongs["track_genre"]=="acoustic"].sort_values(by="popularity",ascending=False).head(50)
Classical=AllSongs[AllSongs["track_genre"]=="classical"].sort_values(by="popularity",ascending=False).head(50)
Jazz=AllSongs[AllSongs["track_genre"]=="jazz"].sort_values(by="popularity",ascending=False).head(50)
Pop=AllSongs[AllSongs["track_genre"]=="pop"].sort_values(by="popularity",ascending=False).head(50)
Rock=AllSongs[AllSongs["track_genre"]=="rock"].sort_values(by="popularity",ascending=False).head(51)
HipHop=AllSongs[AllSongs["track_genre"]=="hip-hop"].sort_values(by="popularity",ascending=False).head(50)
WorldMusic=AllSongs[AllSongs["track_genre"]=="world-music"].sort_values(by="popularity",ascending=False).head(50)

Songs=pd.concat([Acoustic,Classical,Jazz,Pop,Rock,HipHop,WorldMusic])
Songs.reset_index(drop=True,inplace=True)
del Songs["popularity"]
Songs = Songs.drop(200)
Songs.reset_index(drop=True,inplace=True)

names=open("names.txt","r",encoding="utf-8")
names=names.readlines()

#Kişi adları, Kullanıcı adına dönüştürülür(sadece görsellik içindir.)
UserNames=[]
for i in names:
    i=i.strip("\n")
    i=i+"_"+str(random.randint(1,100))
    UserNames.append(i)
UserNames=list(set(UserNames))

#kullanıcı bilgilerinin belli gruplar halinde işlenmesi
genres=["acoustic","classical","jazz","pop","rock","hip-hop","world-music"]
df_list=[]
UsersLogical=[]
for user in range(0,len(UserNames)):
    name=UserNames[user]
    dominant_genre=random.choice(genres)
    song_list=[]
    genre_list=[]
    
    if dominant_genre=="acoustic":
        age=random.randint(18,30)
        musical_knowledge=random.randint(4,8)
            
    if dominant_genre=="classical":
        age=random.randint(12,60)
        musical_knowledge=random.randint(8,10)
        
    if dominant_genre=="jazz":
        age=random.randint(25,40)
        musical_knowledge=random.randint(6,9)
        
    if dominant_genre=="pop":
        age=random.randint(12,25)
        musical_knowledge=random.randint(0,5)
        
    if dominant_genre=="rock":
        age=random.randint(15,20)
        musical_knowledge=random.randint(2,6)
        
    if dominant_genre=="hip-hop":
        age=random.randint(15,35)
        musical_knowledge=random.randint(4,6)
        
    if dominant_genre=="world-music":
        age=random.randint(20,45)
        musical_knowledge=random.randint(5,7)
        
    
    while len(song_list)<24:
        song=random.choice(list(Songs[Songs["track_genre"]==dominant_genre]["Name"]))
        if not song in song_list:
            song_list.append(song)
            
    while len(song_list)<30:
        song=random.choice(list(Songs["Name"]))
        if not song in song_list:
            song_list.append(song)
            
    for i in song_list:
        genre_list.append(list(Songs[Songs["Name"]==i]["track_genre"])[0])

    
    UsersLogical.append([
        age,
        musical_knowledge,
        genre_list.count("acoustic"),
        genre_list.count("classical"),
        genre_list.count("jazz"),
        genre_list.count("pop"),
        genre_list.count("rock"),
        genre_list.count("hip-hop"),
        genre_list.count("world-music")
    ])
    
    song_list.insert(0,name)
    df_list.append(song_list)
    
columns=["Username"]
for i in range(1,31):
    columns.append("Song "+str(i))

#veri setlerinin Müzik İkizi Bulma işlemleri gerçekleştirme üzere kaydedilmesi
    
UsersDf=pd.DataFrame(df_list,columns=columns)
writer=pd.ExcelWriter("UsersAndSongs.xlsx")
UsersDf.to_excel(writer)
writer.save()

UserInfoLogical=pd.DataFrame(UsersLogical,columns=["age","musical_konwledge","acoustic","classical","jazz","pop","rock","hip-hop","world-music"])
writer=pd.ExcelWriter("UsersLogical.xlsx")
UserInfoLogical.to_excel(writer)
writer.save()
```
